```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
result = {
    'ResNet50(imgnet)+ELM' : {
        'Train' : 0.6733,
        'Val' : 0.6184,
        'Test' : 0.6154
    },
     'ResNet50(VGGFace)+ELM' : {
        'Train' : 0.6679,
        'Val' : 0.6208,
        'Test' : 0.6071
    },
     'VGG16(imgnet)+ELM' : {
        'Train' : 0.6873,
        'Val' : 0.6406,
        'Test' : 0.6220
    },
     'VGG16(VGGFace)+ELM' : {
        'Train' : 0.7560,
        'Val' : 0.6968,
        'Test' : 0.7117
    }
}
```

```python
fig, ax = plt.subplots(nrows=1, ncols=len(result), figsize=(20,5), squeeze=False)

for i, (key, value) in enumerate(result.items()):
    df = pd.DataFrame(value, index=[key])
    sns.barplot(data=df, ax=ax[0][i])
    ax[0][0].set_ylabel('Accuracy',fontsize=12)
    ax[0][i].set_title(f'{key} Model',fontsize=14)
    ax[0][i].set_xticklabels(ax[0][i].get_xticklabels(), fontsize=12)

plt.show()
```
